Intel(R) Extension for Scikit-learn is an open source Python package you can use to accelerate your Scikit-learn applications and still have full conformance with all Scikit-Learn APIs and algorithms.

### How it works?

The acceleration is achieved by replacing the stock scikit-learn algorithms with their optimized versions provided by the extension.

One of the ways to do this is by modifying the code:

```python
from sklearnex import patch_sklearn
patch_sklearn()
# After this line, import scikit-learn estimators of your choice
```

To see it in action, let's create some random blobs of data using `scikit-learn` datasets with 6 `centers` and plot it:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

n_clusters = 6
X, y = make_blobs(n_samples=50000, centers=n_clusters, 
                      cluster_std = [0.7, 0.4, 0.5, 0.5, 0.5, 0.5], 
                      random_state=42)

fig, ax = plt.subplots(figsize=(6, 4))
scatter = plt.scatter(X[:, 0], X[:, 1], s=50, c = y, cmap='Set1')
plt.title(f"Random {n_clusters} clusters")
legend1 = ax.legend(*scatter.legend_elements(),title="Clusters")
ax.add_artist(legend1)
```

# Clustering with KMeans 

For simple distributions like the one above, we might try to cluster via Kmeans. Here, although we know the number of clusters is 8, we are choosing a random number for the cluster for kmeans as in practical situations, we won't know the exact number.


````python
from sklearnex import patch_sklearn, unpatch_sklearn
patch_sklearn()  # patch sklearn
from sklearn.cluster import KMeans  # import estimator

n_clusters=5
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)  # kmeans clustering with n_clusters=5

fig, ax = plt.subplots(figsize=(6, 4))
scatter = plt.scatter(X[:, 0], X[:, 1], s=50, c = kmeans.labels_)
plt.title(f"KMeans with n_clusters: {n_clusters}")
legend1 = ax.legend(*scatter.legend_elements(), loc="lower right", title="Clusters")
ax.add_artist(legend1)
plt.show()
````

# Clustering with DBSCAN for more complicated data distributions

We will use DBSCAN from scikit-learn to attemp to identify our clustering.

````python
from sklearnex import patch_sklearn
patch_sklearn()  # patch sklearn
from sklearn.cluster import DBSCAN  # import estimator


clustering = DBSCAN(eps=1.2, min_samples = 10).fit(X)  # DBSCAN fit

fig, ax = plt.subplots(figsize=(6, 4))
plt.title("Using DBSCAN with eps=1.2 and min_samples=10")
scatter = plt.scatter(X[:,0], X[:,1], c = clustering.labels_)
legend1 = ax.legend(*scatter.legend_elements(),title="Clusters")
ax.add_artist(legend1)
plt.show()
````

**Note: DBSCAN Identifies Outliers as class -1**
